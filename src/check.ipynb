```python
import requests
import pandas as pd
import time
import  datetime
from time import sleep
import csv
from sqlalchemy import true
from bs4 import BeautifulSoup as bs

```

```python
df = pd.read_csv('../data/2-8-2022,12:9.csv')
```

```python
df['sold'] = False # is not sold 
a = df['path'][2]
```

```python
i = 0 
b = []
while i in range(len(df)):
    a = df['path'][i]
    response = requests.post(f'https://sa.aqar.fm{a}')
    # response = requests.post(url )
    html = response.content
    soup = bs(html, "lxml")
    i+=1
    try:
        check=soup.find('span', class_='listingScreen_badge__M_A3A')
    
   
        if check.text == ' مغلق':
            df['sold'][i-1] = True
    except:
        try:
            check1=soup.find('div', class_='listingScreen_title__Zy3RL')
            if(check1.text == 'الموقع والخدمات المجاورة' ):
                df['sold'][i-1] = False
            else : 
                    
                df['sold'][i-1] = True 
        except:
            
            df['sold'][i-1] = True
        
        
```
